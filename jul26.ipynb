```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from tabulate import tabulate
```

```python
heart = np.load('data/heart.npy')
heart_rotate = np.load('data/heart_rotate.npy')
heart_scale = np.load('data/heart_scale.npy')
heart_embed = np.load('data/heart_embed.npy')

plt.scatter(heart[:,0],heart[:,1],c = heart[:,1])
plt.title('Heart')
plt.show()
plt.scatter(heart_rotate[:,0],heart_rotate[:,1],c = heart_rotate[:,1])
plt.title('Rotated heart')
plt.show()
plt.scatter(heart_scale[:,0],heart_scale[:,1],c = heart_scale[:,1])
plt.title('Scaled heart')
plt.show()
ax = plt.axes(projection='3d')
ax.scatter3D(heart_embed[:,0], heart_embed[:,1], heart_embed[:,2], c=heart_embed[:,2])
plt.title('Heart embeded into 3D')
plt.show()
```

```python
# heart rotate
path = 'results/heart_heart_rotate/'

# table and plot
data = np.load(path+'heart_heart_rotate.npy')[1:]
print (tabulate(data, headers=["lambda", "gmmd", "mmdx", "mmdy","delta"]))
plt.plot(data[:,0],data[:,1],label = 'gmmd')
plt.plot(data[:,0],data[:,2],label = 'mmdx')
plt.plot(data[:,0],data[:,3],label = 'mmdy')
plt.plot(data[:,0],data[:,4],label = 'delta')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('lambda')
plt.legend()
plt.show()

# epsilon = 0.001~0.512
for i in range(10):
    epsilon = 0.001*(2**i)
    fx = np.load(path+'fx_{e}'.format(e=epsilon)+'.npy')
    gy = np.load(path+'gy_{e}'.format(e=epsilon)+'.npy')
    fgy = np.load(path+'fgy_{e}'.format(e=epsilon)+'.npy')
    gfx = np.load(path+'gfx_{e}'.format(e=epsilon)+'.npy')
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2,2, sharex=True)
    fig.suptitle('lambda = {e}'.format(e=epsilon))
    ax1.scatter(fx[:,0],fx[:,1],c = fx[:,1])
    ax1.title.set_text('f#P')
    ax2.scatter(gy[:,0],gy[:,1],c = gy[:,1])
    ax2.title.set_text('g#Q')
    ax3.scatter(fgy[:,0],fgy[:,1],c = fgy[:,1])
    ax3.title.set_text('fg#Q')
    ax4.scatter(gfx[:,0],gfx[:,1],c = gfx[:,1])
    ax4.title.set_text('gf#P')
    plt.show()
    
```
